## Теория: паттерн Supervisor

**Supervisor** — один агент-координатор (супервизор), который решает, какому подчинённому агенту передать задачу. Подчинённые агенты могут иметь свои инструменты (например, coder пишет код через `write_python_code`, tester запускает тесты через `run_tests`). Супервизор получает ответ от одного агента и может передать результат другому или завершить.

В примере используется `langgraph_supervisor.create_supervisor`: супервизор на базе LLM, подчинённые — ReAct-агенты (create_react_agent) с разными тулами и промптами.
