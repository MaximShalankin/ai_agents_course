## Теория: архитектуры

- **Линейный workflow:** несколько последовательных вызовов LLM (анализ → план → код); состояние передаётся через промпты, без графа.
- **LCEL (LangChain Expression Language):** цепочка runnables (retriever | prompt | llm | parser); композиция через `|`, удобно для пайплайнов без ветвлений.
- **ReAct:** цикл «мысль → действие (tool) → наблюдение»; агент решает, какой tool вызвать и когда остановиться.
- **LangGraph:** явный граф состояний (StateGraph), узлы — LLM или ToolNode, рёбра — переходы; поддержка циклов и условных переходов (tools_condition).
