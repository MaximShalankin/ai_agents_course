## Теория: метрики оценки кодогенерации

- **Синтаксическая корректность:** из ответа LLM извлекают блок кода (например, из ```python ... ```), проверяют через `ast.parse`; доля сэмплов с валидным синтаксисом — базовая метрика.
- **Прохождение тестов (test pass@k):** для каждого сэмпла HumanEval есть эталонный тест; подставляют сгенерированное решение и запускают тест; считают долю решённых задач (или pass@1).
В примере используется только проверка синтаксиса; тесты HumanEval можно добавить по той же схеме (exec + check).
